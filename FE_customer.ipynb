## Feature Encoding
### 1.1 Label encoding

```python
encoder = LabelEncoder()

encoder.fit(df['State'])

df['state_encoded'] = encoder.transform(df['State'])

df
```

```python
encoder.classes_
```

### 1.2 Ordinal encoding

```python
df.VehicleClass.unique()
```

```python
VehicleClass_categories = ['SUV', 'Four-Door Car', 'Luxury SUV', 'Luxury Car', 'Sports Car', 'Two-Door Car' ]
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories = [VehicleClass_categories])

df['VehicleClass'] = encoder.fit_transform(df[['VehicleClass']])

df
```

```python
df = df.drop('State' , axis = 1)
```

```python
df
```

### 1.3 One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

obj = OneHotEncoder(sparse_output = False, dtype = int)

new_col = obj.fit_transform(df[['Gender']])

print(type(new_col))
```

```python
encoded_df = pd.DataFrame(new_col, columns = obj.get_feature_names_out(['Gender']))

encoded_df
```

```python
df = pd.concat([df,encoded_df], axis = 1)

df
```

```python
df = df.drop('Gender', axis = 1)
df
```

## 2. Chi-Square test

```python
from scipy.stats import chi2_contingency
```

```python
contigency_matrix = pd.crosstab(df['Policy'], df['PolicyType'])
```

```python
_ , p_value, _ , _ = chi2_contingency(contigency_matrix)
```

```python
p_value
```

```python
# p < 0.05 : their is significant relation between 2 variables, Null hypothesis fails
# p > 0.05 : their is no significant relation between 2 variables, Null hypothesis don't fail
```

## Feature Balancing

### 3.1 Oversampling

```python
df.Response.value_counts().plot(kind = 'pie', autopct = "%0.0f%%")
```

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
```

```python
x = df.drop('Response' , axis =1)
y = df.loc[: , ['Response']]
```

```python
obj_ros = RandomOverSampler(sampling_strategy = 'minority')
x_ros, y_ros = obj_ros.fit_resample(x,y)
```

```python
y_ros.value_counts()
```

```python
y.value_counts()
```

```python
obj_smote = SMOTE(sampling_strategy = .5 )
x_smote, y_smote = obj_smote.fit_resample(x,y)
```

```python
y_smote.value_counts()
```

## 3.2 Undersampling :decrease the frequency of the majority class
### 3.2.1 Random Under sampler
### 3.2.2 Edited Nearest Neighbour

```python
from imblearn.under_sampling import RandomUnderSampler

obj_rus = RandomUnderSampler(sampling_strategy = 'not minority')
x_rus, y_rus = obj_rus.fit_resample(x,y)
```

```python
y_rus.value_counts()
```

```python
from imblearn.under_sampling import EditedNearestNeighbours

obj_enn = EditedNearestNeighbours(n_neighbors = 5)
x_enn, y_enn = obj_enn.fit_resample(x,y)
```

```python
y_enn.value_counts()
```
